```python
import pandas as pd
df=pd.read_excel("c:/Users/HP/Downloads/shipment_dataset_10000.xlsx")
df
```

```python
# remove duplicates 
df.drop_duplicates()
```

```python
# outlier anomaly
outliers = df[df['shipping_distance_km'] > 20000]
print("Outliers:\n", outliers)
# Replace with the median value
median_value = df['shipping_distance_km'].median()
df.loc[df['shipping_distance_km'] > 20000, 'shipping_distance_km'] = median_value
df
```

```python
# datatype mismatch
df['supplier_rating'] = df['supplier_rating'].astype(str).str.replace('"', '')
df['supplier_rating'] = pd.to_numeric(df['supplier_rating'], errors='coerce').astype(float)
df
```

```python
# Missing values
rating = df['supplier_rating'].mean()
df.loc[:, 'supplier_rating'] = df['supplier_rating'].fillna(rating)
df['supplier_rating'] = df['supplier_rating'].round(1)
df
```

```python
# invalid values
df.loc[df['supplier_lead_time'] < 0, 'supplier_lead_time'] = \
     df['supplier_lead_time'].abs()
df
```

```python
df.to_excel("output.xlsx", index=False)
```
